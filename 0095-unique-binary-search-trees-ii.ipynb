```python
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def generateTrees(n):
    """
    :type n: int
    :rtype: List[TreeNode]
    """
    if n == 0:
        return []

    nums = [i for i in range(1,n+1)]
    
    def helper(nums):

        if not nums:
            return [None]
        
        res = []
        for i, num in enumerate(nums):
            for l in helper(nums[:i]):
                for r in helper(nums[i+1:]):
                    node = TreeNode(num)
                    node.left = l
                    node.right = r
                    res.append(node)
        return res

    return helper(nums)
```

```python
n = 3
generateTrees(n)
```
