### 79. 单词搜索
给定一个 m x n 二维字符网格 board 和一个字符串单词 word 。如果 word 存在于网格中，返回 true ；否则，返回 false 。

单词必须按照字母顺序，通过相邻的单元格内的字母构成，其中“相邻”单元格是那些水平相邻或垂直相邻的单元格。同一个单元格内的字母不允许被重复使用。

示例 1：



输入：board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]], word = "ABCCED"
输出：true

https://leetcode.cn/problems/word-search

**思路**：回溯法

控制棋盘ij和字符idx，超过边界或者字符不相等返回False；遍历到某个直接修改成“*”，进行4种情况的递归；棋盘还原(为了循环时候棋盘保持一致)；返回四种情况的或布尔值。

```python
def exist(board, word):
    """
    :type board: List[List[str]]
    :type word: str
    :rtype: bool
    """

    def backtrack(word,i, j, board, idx):

        if i<0 or i>=len(board) or j<0 or j >=len(board[0]) or board[i][j]!=word[idx]:
            return False

        if idx == len(word)-1:
            return True

        board[i][j] = '*'

        r1 = backtrack(word,i+1, j, board,idx+1)
        r2 = backtrack(word,i-1, j, board,idx+1)
        r3 = backtrack(word,i, j+1, board,idx+1)
        r4 = backtrack(word,i, j-1, board,idx+1)
        res = r1 or r2 or r3 or r4

        board[i][j] = word[idx]

        return res


    return any([backtrack(word, i, j, board, 0) for i in range(len(board)) for j in range(len(board[0]))])
```

```python
board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]]
word = "BCCED" 

def backtrack(word,i, j, board, idx):
    
    if i<0 or i>=len(board) or j<0 or j >=len(board[0]) or board[i][j]!=word[idx]:
        return False

    if idx == len(word)-1:
        return True

    board[i][j] = '*'
    
    r1 = backtrack(word,i+1, j, board,idx+1)
    r2 = backtrack(word,i-1, j, board,idx+1)
    r3 = backtrack(word,i, j+1, board,idx+1)
    r4 = backtrack(word,i, j-1, board,idx+1)
    res = r1 or r2 or r3 or r4
    
    board[i][j] = word[idx]
        
    return res


any([backtrack(word, i, j, board, 0) for i in range(len(board)) for j in range(len(board[0]))])
```
