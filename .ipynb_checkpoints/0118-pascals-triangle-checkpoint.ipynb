****

=j-1j

```python
def generate(numRows):
    """
    :type numRows: int
    :rtype: List[List[int]]
    """
    res = [[1]]
    for i in range(1, numRows):
        tmp = [1]
        for j in range(1, i):
            tmp.append(res[-1][j-1]+res[-1][j])
        tmp.append(1)
        res.append(tmp)

    return res
```

```python
numRows = 5
generate(numRows)
```
