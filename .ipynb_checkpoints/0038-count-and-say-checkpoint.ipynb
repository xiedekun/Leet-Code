*****

```python
def countAndSay(n):
    """
    :type n: int
    :rtype: str
    """
    if n==1:
        return '1'
    
    s = countAndSay(n-1) + '*'
    res, cnt = '',1
    for i in range(len(s) -1):
        if s[i] == s[i+1]:
            cnt+=1
        else:
            res += str(cnt) + str(s[i])
            cnt = 1
    return res
```

```python
countAndSay(3)
```
