```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
def partition(head, x):
    """
    :type head: ListNode
    :type x: int
    :rtype: ListNode
    """
    if not head:
        return 
    
    low = search = dummyhead = ListNode(-float('inf'), head)

    while low.next and low.next.val < x:
        low = low.next
        search = search.next

    while search and search.next:
        
        if search.next.val < x:
            temp = search.next
            search.next = search.next.next
            temp.next = low.next
            low.next = temp
            low = low.next
        else:
            search = search.next

    return dummyhead.next
        
```

```python
head = ListNode(1,ListNode(4,ListNode(3,ListNode(0,ListNode(2,ListNode(5,ListNode(2)))))))
x = 3
# head = ListNode(2, ListNode(1))

head = partition(head, x)
while head:
    print(head.val)
    head = head.next
```
