### 82. 删除排序链表中的重复元素 II
给定一个已排序的链表的头 head ， 删除原始链表中所有重复数字的节点，只留下不同的数字 。返回 已排序的链表 。

 

示例 1：



输入：head = [1,2,3,3,4,4,5]
输出：[1,2,5]

https://leetcode.cn/problems/remove-duplicates-from-sorted-list-ii/

**思路**：快慢双指针＋慢指针的前一个

快的先探路，当快慢元素始终不同，上一个指针指向满指针；如果期间有相同元素，上一个指针的下一个指向快指针；慢指针指到快指针，快指针下一个。

```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
def deleteDuplicates(head):
    """
    :type head: ListNode
    :rtype: ListNode
    """
    
    if not head:
        return head
    
    lastnode = None

    dummyhead = slow = fast = ListNode(None)

    dummyhead.next = head
    fast = fast.next

    while fast:
        flag = False
        while fast and fast.val == slow.val:
            fast = fast.next
            flag = True

        if flag:
            lastnode.next = fast
        else:
            lastnode = slow

        slow = fast

        if fast:
            fast = fast.next
            
    return dummyhead.next
```

```python
head = ListNode(1,ListNode(2,ListNode(3,ListNode(3,ListNode(4,ListNode(4,ListNode(5)))))))

node = deleteDuplicates(head)


        

while node:
    print(node.val)
    node = node.next
```
