### 24. 两两交换链表中的节点
给你一个链表，两两交换其中相邻的节点，并返回交换后链表的头节点。你必须在不修改节点内部的值的情况下完成本题（即，只能进行节点交换）。


输入：head = [1,2,3,4]

输出：[2,1,4,3]

https://leetcode.cn/problems/swap-nodes-in-pairs/comments/

```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
l = ListNode(1, ListNode(2, ListNode(3, ListNode(4))))
```

**思路**：递归：

```python
def swapPairs(head):
    """
    :type head: ListNode
    :rtype: ListNode
    """
    if head == None or head.next==None:
        return head
    
    temp = head.next
    head.next = swapPairs(head.next.next)
    temp.next = head
    
    return temp

l = ListNode(1, ListNode(2, ListNode(3, ListNode(4))))
swapPairs(l).next.val
```

**思路**：两两入新的链表

```python
def swapPairs(head):
    """
    :type head: ListNode
    :rtype: ListNode
    """
    res = dh = ListNode(None)
    node = head
    
    while node:
        if node.next:  
            dh.next = ListNode(node.next.val)
            dh.next.next = ListNode(node.val)
        else:
            dh.next = ListNode(node.val)
            return res.next
        
        node = node.next.next
        dh = dh.next.next

    return res.next

swapPairs(l).next.val
```
