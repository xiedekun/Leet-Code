### 86. 分隔链表
给你一个链表的头节点 head 和一个特定值 x ，请你对链表进行分隔，使得所有 小于 x 的节点都出现在 大于或等于 x 的节点之前。

你应当 保留 两个分区中每个节点的初始相对位置。

 

示例 1：



输入：head = [1,4,3,2,5,2], x = 3

输出：[1,2,2,4,3,5]

https://leetcode.cn/problems/partition-list/

**思路**：双指针+dummyhead

使用dummyhead是防止开头就是较大值的情况。

开始low指针和search指针先找到最后一个小于target的值，再用search指针往下搜索，找到后一个小于的接在low指针之后并交换节点；如果没发生交换search指针指向下一个。

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
def partition(head, x):
    """
    :type head: ListNode
    :type x: int
    :rtype: ListNode
    """
    if not head:
        return 
    
    low = search = dummyhead = ListNode(-float('inf'), head)

    while low.next and low.next.val < x:
        low = low.next
        search = search.next

    while search and search.next:
        
        if search.next.val < x:
            temp = search.next
            search.next = search.next.next
            temp.next = low.next
            low.next = temp
            low = low.next
        else:
            search = search.next

    return dummyhead.next
        
```

```python
head = ListNode(1,ListNode(4,ListNode(3,ListNode(0,ListNode(2,ListNode(5,ListNode(2)))))))
x = 3
# head = ListNode(2, ListNode(1))

head = partition(head, x)
while head:
    print(head.val)
    head = head.next
```
