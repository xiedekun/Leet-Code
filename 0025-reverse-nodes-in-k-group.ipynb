```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
def reverseKGroup(head, k):
    """
    :type head: ListNode
    :type k: int
    :rtype: ListNode
    """
    h = head
    for i in range(k):
        if h == None:
            return head
        h = h.next
    
    pre = None
    nxt = None
    first = None
    for i in range(k):
        nxt = head.next
        if i==0:
            first = head
        head.next = pre
        pre = head
        head = nxt
    first.next = reverseKGroup(head, k)
    
    return pre
    

l = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))
l = reverse(l, 4)

while l:
    print(l.val)
    l = l.next

        
```
