### 21. 合并两个有序链表
将两个升序链表合并为一个新的 升序 链表并返回。新链表是通过拼接给定的两个链表的所有节点组成的。 



输入：l1 = [1,2,4], l2 = [1,3,4]

输出：[1,1,2,3,4,4]

**思路**：归并+dummyhead

```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
l1 = ListNode(1, ListNode(2, ListNode(4)))
l2 = ListNode(1, ListNode(3, ListNode(4)))

res = head = ListNode(None)

while l1 or l2:
    
    if l1 == None:
        head.next = ListNode(l2.val)
        l2 = l2.next
        head = head.next
        continue
    elif l2 == None:
        head.next = ListNode(l1.val)
        l1 = l1.next
        head = head.next
        continue
        
        
    v1 = l1.val
    v2 = l2.val
    
    if v1<=v2:
        head.next = ListNode(v1)
        l1 = l1.next
    else:
        head.next = ListNode(v2)
        l2 = l2.next
    head = head.next
    
res.next.next.next.val
```
