```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def sortedListToBST(head):
    """
    :type head: Optional[ListNode]
    :rtype: Optional[TreeNode]
    """
    if not head:
        return 
    
    if not head.next:
        return TreeNode(head.val)
    
    fast = pre = mid = head

    mid = mid.next
    fast = fast.next.next

    while fast and fast.next:       
        fast = fast.next.next
        pre = pre.next
        mid = mid.next

    pre.next = None
    
    root = TreeNode(mid.val)
    root.left = sortedListToBST(head)
    root.right = sortedListToBST(mid.next)
    
    return root
```

```python
head = [-10,-3,0,5,9]
h = p = ListNode(None)

for num in head:
    p.next = ListNode(num)
    p = p.next
head = h.next

sortedListToBST(head).left.val
```
