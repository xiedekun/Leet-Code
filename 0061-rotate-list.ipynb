### 61. 旋转链表
给你一个链表的头节点 head ，旋转链表，将链表每个节点向右移动 k 个位置。

 

示例 1：



输入：head = [1,2,3,4,5], k = 2

输出：[4,5,1,2,3]

**思路**：先统计链表长度；根据余数移动快慢指针。

```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
def rotateRight(head, k):
    """
    :type head: ListNode
    :type k: int
    :rtype: ListNode
    """
    if not head:
        return 
    cnt = 1
    
    point = fast = slow = head

    while point.next:
        cnt+=1
        point = point.next
    
#     print(k%cnt)
    if k%cnt == 0:
        return head
    
    for _ in range(k%cnt):
        fast = fast.next
        
    while fast.next:
        fast = fast.next
        slow = slow.next

    movenode = slow.next
    slow.next = None
    fast.next = head
    
    return movenode
```

```python
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))
k = 5
rotateRight(head, k).val
```
